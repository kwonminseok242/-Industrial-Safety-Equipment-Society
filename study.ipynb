## 1단계. AI/딥러닝 기본 개념 (분류 vs 탐지 vs 세그멘테이션)

- **분류(Classification)**  
  - 한 장 이미지 → 클래스 하나 예측  
  - 예: X-ray → 정상 vs 폐렴

- **탐지(Detection)**  
  - 한 장 이미지 → 여러 객체 **위치(박스) + 종류(클래스)**  
  - 예: 건물 사진 → 균열, 박리, 백태 위치 탐지

- **세그멘테이션(Segmentation)**  
  - 한 장 이미지 → 픽셀 단위로 어떤 클래스인지 구분  
  - 예: 균열 라인 전체 모양 추출

👉 요약:  
- 분류 = 무엇  
- 탐지 = 어디 + 무엇  
- 세그멘테이션 = 어디 + 무엇 + 픽셀 단위로 정확히

---

## 2단계. 성능 지표 (mAP, Precision, Recall)

- **Precision (정밀도)**: 탐지한 것 중 맞은 비율  
- **Recall (재현율)**: 실제 있는 것 중 내가 찾은 비율  
- **mAP (mean Average Precision)**: IoU 여러 기준(0.5~0.95)에서 평균 낸 점수  

👉 예시:  
- Precision ↑, Recall ↓ → 맞을 때는 잘 맞추지만 놓침이 많음  
- Precision ↓, Recall ↑ → 다 찾긴 하지만 오탐이 많음  

---

## 3단계. YOLO 이미지 처리 방식

1. 이미지를 **grid**로 나눔  
2. 각 셀에서 박스 + 클래스 예측  
3. 겹친 박스는 **NMS(비최대 억제)**로 정리  
4. 결과: 이미지 속 여러 객체의 위치와 클래스 출력

👉 핵심: **You Only Look Once** → 이미지를 한 번 보고 다 예측

---

## 4단계. Ultralytics YOLO 명령어 구조

- **학습(train)**  
  ```bash
  yolo task=detect mode=train model=yolov8n.pt data=configs/data/defects.yaml epochs=100
  ```

- **검증(val)**  
  ```bash
  yolo task=detect mode=val model=best.pt data=configs/data/defects.yaml
  ```

- **추론(predict)**  
  ```bash
  yolo task=detect mode=predict model=best.pt source=data/images/val
  ```

👉 핵심: `task=detect | segment`, `mode=train | val | predict`

---

## 5단계. YOLO 라벨 포맷(txt)

- 포맷:  
  ```text
  class x_center y_center width height
  ```

- 모든 좌표는 **0~1 정규화**

- 예시 (`dog.jpg` → `dog.txt`):  
  ```text
  0 0.5 0.5 0.2 0.3
  ```
  → 클래스 0, 중심(0.5, 0.5), 크기(0.2, 0.3)

---

## 6단계. 정리

- YOLO = 탐지 모델 → 위치+종류 동시에  
- 성능 지표 = Precision / Recall / mAP  
- CLI 기본 = train / val / predict  
- 라벨 txt = 클래스 ID + 정규화 좌표
